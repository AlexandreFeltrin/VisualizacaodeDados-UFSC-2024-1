```python
media_notas_geral=dfNotas['Nota'].mean()

media_notas_geral
```

```python
notas10_15=dfNotas[(dfNotas['Idade'] >= 10) & (dfNotas['Idade'] <= 15)]


notas10_15
```

```python
print("A média de notas da Faixa etária é 10 a 15 anos" )
notas10_15['Nota'].mean()
```
